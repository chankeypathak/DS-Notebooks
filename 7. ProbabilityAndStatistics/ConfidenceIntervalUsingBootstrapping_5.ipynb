# Confidence Interval Using Bootstrapping

Through this technique CI can be calculated for Median,variance,std div, 90th percentile

## CI for Median

Similarly instead of calculating Median of sample if we calculate std dev it would be CI of std Dev

# Empirical Bootstrap Based Confidence Interval

```python
import numpy
from pandas import read_csv
from sklearn.utils import resample
from sklearn.metrics import accuracy_score
from matplotlib import pyplot
```

```python
# data set

x = numpy.array([180,162,158,172,168,150,171,183,165,176])

#Configuring bootstarp
n_iterations = 1000
n_size = int(len(x))

#run bootstrap
medians = list()

# creating 1000 samples with repatition allowed 
# and calculating medians of these samples 
for i in range(n_iterations):
    s = resample(x, n_samples = n_size)
    m = numpy.median(s)
    medians.append(m)

# plot scores i.e plotting histogram of medians
pyplot.hist(medians)
pyplot.show()

# Confidence Intervals
alpha = 0.95; # 95%
p = ((1.0 - alpha)/2.0) * 100; # its 0.5/2 i.e 2.5% -> 2*2.5 is rest 5%
lower = numpy.percentile(medians,p); #2.5%

# 95% + 2.5% -> 97.5%
p = (alpha+((1.0 - alpha)/2.0)) * 100; 
upper = numpy.percentile(medians,p); # 97.5%

print("%f Confidence Interval for median is %f and %f" %(alpha*100,lower,upper))
```

Note: Lower is 2.5% Upper is 97.5% interval is 97.5% - 2.5% = 95%
