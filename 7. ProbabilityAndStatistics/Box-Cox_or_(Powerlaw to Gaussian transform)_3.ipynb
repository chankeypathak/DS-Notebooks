# Box-Cox/Power Transform

Its nothing but converting PowerLaw(80-20 based distribution) to Gaussian Distribution.

Explanation:<br>
Consider a random var 'X' it is powerlaw/perito distributed Which is Converted to 'Y' Gaussian Distributed by applying Box-Cox Transform.<br>

Box-Cox will calculate 'Lambda' if Lambda=0 -> Apply ln(xi) <br>else (pow(xi,lambda)-1)/X <br>

This above Calculation is done by Box-Cox Function.

```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
```

Generating rand vars from a non-normal/non gaussian distribution and make a probability plot(Q-Q plot) for it, to show it is non-normal in the tails of plot.

```python
fig = plt.figure()
ax1 = fig.add_subplot(211)# 2 rows 1 col 1st sub plot

# here we are creating loggamma distributed(it follows 80-20 rule i.e powe law)
# rand vars 
x = stats.loggamma.rvs(5,size=500) +5

# Q-Q plot 
prob = stats.probplot(x,dist=stats.norm,plot=ax1)
plt.xlabel('Std Normal Distribution')
plt.ylabel('Loggamma Distribution/Q-Q')

ax2 = fig.add_subplot(212)

#converting 'x'(loggamma/power law dist) to 'y'(Normal distribution)
# here '_' is nothing but the lambda val returned by the function
# lambda is log of x
y, lambda_val = stats.boxcox(x);
prob = stats.probplot(y,dist=stats.norm,plot=ax2)
ax2.set_title('ProbPlot After Box-cox Transform')
plt.show()
```

Observation: In the first plot we can see that some values are deviated i.e the dont follow normal distribution, In the second plot after boxcox transform there is much less deviation i.e they follow Normal Distribution.

Note: Here we used Q-Q plot because Q-Q plot decides if both x,y follow same distribution or not.
