# KS-Test(Kolmogorov-Smirnov Test)

It's bascially used to determine if Two Rand Vars X & Y are having same distribution or not

<img src="capture.png">

Observation:<br>
In the above plot there are CDF's of two random variables X1 & X2, Here the test Static 'D' is nothing but Maximum of difference between the distance of 2 CDF's.
<br>

If 'D' is 0 then Assumed null hypothesis h0 is true, if 'D'>0 then h0 is false.

<img src="kstest1.png">

if D >0.047 then H0 is rejected, o.047 is calculated with the above formula,
alpha is level of significance like 10% 5% 2.5%....

```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# generating a gaussian r.v X

x = stats.norm.rvs(size=1000);
sns.set_style('whitegrid')
sns.kdeplot(x,bw=0.5) #kernal density estimation plot/ pdf
plt.xlabel("Normal Distribution")
plt.ylabel("PDF of X")
plt.show()
```

## KS TEST

```python
#comparing x with normal distribution, this second parm can be any rand var
d,alpha=stats.kstest(x,'norm') 
print("D val is: ",d,", alpha/ level of significance: ",alpha)
```

as alpha is very high than D its normally distributed

#### Another example

```python
# y is continuous uniform distribution(0,1)
y = np.random.uniform(0,1,1000)
sns.kdeplot(y,bw=0.1)
plt.show()
```

```python
d,alpha=stats.kstest(y,'norm') 
print("D val is: ",d,", alpha/ level of significance: ",alpha)
```

Knowingly we are comparing Normal, uniform distributions To demonstrate KS test functionality as they both are different pvalue/alpha is 0(i.e D>alpha) that indicates They dont follow same distribution
