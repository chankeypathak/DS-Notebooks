# Q-Q Plot
(Quantile Quantile plot)
This plot is used to answer:<br>

1) Is Random Var 'X' is normally/gaussian distributed?<br>
2) If there are 2 rand vars X,Y, does they have same/similar distribution?
<br>
### Limitations:<br>

When Sample size is less its some what hard to predict weather its normally distributed or not.

```python
import numpy as np
import pylab; #its used to plot a graph
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# np.random.normal -> creates random values/samples which are normally distributed
# loc -> mean
# scale -> std deviation
# size -> no.of values/samples

#N(0,1) mean 0, std dev 1.
std_normal = np.random.normal(loc = 0,scale = 1,size = 1000)
std_normal.size
```

```python
# 0-100 percentiles of std_normal

for i in range(0,101,4): #step size 4
    print(i, np.percentile(std_normal,i))

# or
    
#l = [{i:np.percentile(std_normal,i)} for i in range(0,101)]
#l
```

```python
# generating 100 samples from N(20,5) ->mean 20 std dev 5

measurements = np.random.normal(loc = 20, scale = 5, size=100)

stats.probplot(measurements,dist='norm',plot=pylab)
pylab.show()
```

#### Limitations:<br>

When Sample size is less its some what hard to predict weather its normally distributed or not.

```python
measurements = np.random.normal(loc = 20, scale = 5, size=10000)

stats.probplot(measurements,dist='norm',plot=pylab)
pylab.show()
```

Note:<br>
x axis is std normal distribution i.e N(0,1)
y axis is N(20,5)

### Plotting uniform vs normal distribution

```python
measurement = np.random.uniform(low = -1, high=1,size=10000)
stats.probplot(measurement,dist='norm',plot=pylab)
pylab.show()
```

Y- axis is uniform distribution, X asis is Normal Distribution
